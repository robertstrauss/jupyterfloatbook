```python
def update(string):    
    # calculate derivatives
    dstring = np.gradient(string, axis=0)
    ddstring = np.gradient(dstring, axis=0)
    
    # elasticity between beads, pushing/pulling them to be unit distance apart
#     distance = dstring
    elastic = - ( dstring - dstring/np.linalg.norm(dstring, axis=1)[:,np.newaxis] )
#     elastic = elastic*np.linalg.norm(elastic,axis=1)[:,np.newaxis]
    string = string + elastic*dt
    
    # minimize curvature
    string = string + np.sign(ddstring)*dt
    
    # apply boundary conditions
    for ind in boundaryconds:
        string[ind] = boundaryconds[ind]
    
    return string
```

```python
plt.ion()
fig, ax = plt.subplots()
line = ax.plot(*string.T)
# bend = ax.plot(*(string+ddstring).T)
# plt.axis([0, 10, 0, 1])
# plt.figure()
def animate(i):
    global string
    for ii in range(200):
        string = update(string)
    print('frame');
    line.set_xdata(string[:,0])
    line.set_ydata(string[:,1])
    
#     fig.canvas_draw()
#     line.set_offsets(string)
#     bend.set_offsets((string+ddstring))

#     fig.canvas.flush_events()
    fig.canvas.draw_idle()
#     string = string1
#     time.sleep(0.2)
#     plt.scatter(string[:,0], string[:,1])
#     ax.scatter(string[:,0], string[:,1])
#     plt.pause(0.1)

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=100000, interval=20, repeat=False)
plt.show()
```
