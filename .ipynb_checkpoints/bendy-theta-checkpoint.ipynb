# create a random path (inital condition) using a fixed-step random walk

## it will go in random directions, so we will create it by randomizing the derivative then integrating

```python
length = 500
```

```python
theta = np.random.rand(length)*2*np.pi
```

```python
plt.figure()
plt.plot(theta)
```

```python
def anglestopath(theta):
    """
        integrate a list of angles into a path
    """
    return np.cumsum(
            np.array([np.cos(theta), np.sin(theta)])
        , axis=1)
```

```python
strip = anglestopath(theta)
```

```python
plt.figure()
plt.plot(*strip)
```

# create simulation

```python
dt = 0.1
boundaryconds = {
    0: np.array([0, 0]),
#     50: np.array([0, 50]),
#     -1: np.array([70, -10])
    -1: np.array([length/4, length/4])
}
```

```python
def update():
    global theta
    
    # calculate bend between each point
    bend = np.diff(theta)
    
    # each point tries to reduce bend
    theta[1:] -= bend/2 * dt
    theta[:-1] += bend/2 * dt
    
#     # apply boundary conditions
#     for ind in boundaryconds:
#         string[ind] = boundaryconds[ind]
```

# animate and run simulation

```python
# plt.ion()
fig, ax = plt.subplots()
line = ax.plot(*strip)

for i in range(200):
    for ii in range(20):
        update()
    print('fucking hell')
    line.set_xdata(strip[0])
    line.set_ydata(strip[1])
#     fig.canvas.draw_idle()
#     return [line]
# somename = matplotlib.animation.FuncAnimation(fig, somefunctionname, frames=100, interval=200, repeat=False)
```
